One of the most important parts of modeling and data analysis in neuroscience involves looking at raw data, preprocessed data, etc. I've found that MNE is really useful in doing this, with its raw data view, PSD viewer, etc.

However, one challenge I've faced in using MNE for ECoG is that grid placement is unique for each person, not standardized like an MEG or EEG setup. MNE has great functions for quickly plotting lots of data at various points on an EEG grid, but it's not as straightforward to build something custom for your particular setup.

Luckily, there _is_ a way to do this:

```python
# First we list what's in our HDF file
data_path = '../data/ecog_test_data.h5'

with tbl.open_file(data_path, 'r') as f:
    print(f.list_nodes('/'))
```

First load the data. The "ecog" object is a collection of an electrophysiological signal, as well as metadata about time during the collection of that signal.

For more information about what's inside, check out the tutorial on creating events and epochs objects [here](http://nbviewer.ipython.org/github/choldgraf/write-ups/blob/master/mne/mne_events_epochs.ipynb)

```python
# First, we'll load the data just so that we have metadata about electrode placement and trial info.
ecog = ecg.struct.Ecog.from_hdf(data_path, 'preproc')

# We'll create a few dummy channels, and then convert to MNE format
n_dummies = 64
dummy_data = np.random.randn(ecog.signal.shape[0], n_dummies)
ecog.signal = pd.DataFrame(dummy_data, columns=range(1, n_dummies+1))

ecog_mne = ecog.to_mne(ch_types='eeg', stims=ecog.signalmetadata['t_type'])
```

However, we'd like something that conforms to the dimensions of an ecog grid. Fortunately, I have marked the X/Y positions of several electrodes on a sample grid. Here's what the data looks like:

```python
ecog.colmetadata.head()
```

We can use this data to create a custom layout in MNE, then make pretty plots from this.

```python
# We need to use the layout module of MNE
from mne.channels.layout import Layout
```

## Layout takes these parameters

```python
# Box: a 4-item tuple that defines the min/max of X and the min/max of Y that will define the figure to plot.
#      We'll define it as bounded roughly by 0 and 1 on all sides to make things easier (plus some padding)
box = (-.1, 1.1, -.1, 1.1)
```

```python
# Pos: An N x 4 array where N is the number of signals. Each item has the X, Y position of the axis for that signal, 
#      as well as the height and width of the signal.
#      We'll normalize our X/Y values from before so that they're between 0 and 1
xy = ecog.colmetadata[['x_2d', 'y_2d']].dropna()
n_xy = xy.shape[0]

def normalize(arr):
    arr -= arr.min()
    arr /= (arr.max() - arr.min())
    return arr
xy = xy.apply(normalize)
xy['y_2d'] = 1 - xy['y_2d']  # Need to flip the y-axis for plotting

# Quickly define width and height, then expand to the right number of 
w, h = [.05, .07]
w, h = [np.tile(i, xy.shape[0]) for i in [w, h]]

# Now create the pos array
pos = np.hstack([xy, w[:, None], h[:, None]])
```

```python
# ch_names, ch_indices, name: These are there to make sure that each location corresponds to an electrode that exists in our
# MNE object.
ch_names = ecog_mne.ch_names
ch_indices = range(n_xy)
name = 'ecog_layout'
```

```python
# Now we can create our layout
layout = Layout(box, pos, ch_names, ch_indices, name)
```

Now that we've got a layout, we can use it to plot channel information at its location on the grid.

```python
events = mne.find_events(ecog_mne, 't_type')
epochs = mne.Epochs(ecog_mne, events, {'bef': 2, 'aft': 3}, -.2, 1, picks=ch_indices)
```

```python
# If we want an interactive map
# %matplotlib qt

# If we want it plotted inline but static
%matplotlib inline
```

```python
f = mne.viz.plot_topo_image_epochs(epochs, layout=layout,
                                   scalings={'eeg': 2}, layout_scale=.7,
                                   title='Custom ECoG layout')
```

And there you have a custom ECoG topographic layout!

## Afterward / Caveats

1. A few hiccups I ran into figuring this out: for one, when you define box / pos / etc, I could never get it to work when I used anything other than a box bounded around 0 and 1, with all my axis positions lying between 0 and 1 as well. So for me, the normalization step was necessary.
2. Sometimes the layout doesn't all plot in the same window. For this, tweaking layout_scale helped a lot.
3. Any time you plot using ECoG, the scaling of the electrodes will usually be way off compared with what MNE expects with the "scalings" parameter. You should define your own scaling to make plotting look right.
