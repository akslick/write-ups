```python
import numpy as np
import pandas as pd
import scipy as sp
import tables as tbl
import ecog as ecg
import mne
```

```python
data_path = '../data/ecog_test_data.h5'
```

```python
# First we list what's in the HDF file
with tbl.open_file(data_path, 'r') as f:
    print(f.list_nodes('/'))
```

First load the data, the "ecog" object is a collection of an electrophysiological signal, as well as metadata about time during the collection of that signal

```python
ecog = ecg.struct.Ecog.from_hdf(data_path, 'preproc')
```

```python
ecog.signal.iloc[:, 0].head(10000).plot()
```

We've also got time information about this data:

```python
ecog.signalmetadata
```

MNE packages metadata along with the signal itself, so we'll need to append the time information to our signal, and then label it accordingly in MNE

```python
signal_to_convert = ecog.signal
signal_to_convert.head()
```

```python
signal_to_convert['t_type'] = ecog.signalmetadata['t_type']
ch_types.append('misc')  # How we define stim channels in mne
```

First, we'll create an MNE object from our raw data. MNE Requires that we give it an array with values, as well as an "info" object that contains information about the data

```python
# MNE Requres channel names to be strings
ch_names = list(signal_to_convert.columns.astype(str))
sfreq = ecog.Fs
ch_types = ['eeg'] * len(ch_names)
ch_types[-1] = 'misc'  # For the stim channel
```

```python
info = mne.create_info(ch_names, sfreq, ch_types)
```

Now we'll create our MNE object

```python
data_mne = mne.io.RawArray(signal_to_convert.values.T, info)  # want signals x times
```

Now we have an MNE object that we can do lots of interesting things with. For example, I can look at the power spectrum of my signal:

```python
_ = data_mne.plot_psds()
```

## Separating out trials

MNE can use the stim channel of our data to pull out trials. We must use a function for doing so, and point it to the correct stim channel.

```python
trials = mne.find_events(data_mne, 't_type')
```

Then, we can use the trials information to separate out our data

```python
epochs = mne.Epochs(data_mne, trials, {'second': 2, 'third': 3}, -.2, 2)
```

```python
epochs_df = epochs.as_data_frame()
```

```python
epochs_df.query('condition == "second"').mean(level=('time')).plot(legend=False)
```

We can use the epochs object to do some quick time-frequency decompositions

```python
freqs_tfr = np.logspace(0, np.log10(150), 20)
epochs_tfr_power, epochs_tfr_coh = mne.time_frequency.tfr_morlet(epochs, freqs_tfr, 1)
```

```python
plot_elec = range(3)
epochs_tfr_power.plot(plot_elec)
```

```python
epochs_tfr_coh.plot(plot_elec)
```
